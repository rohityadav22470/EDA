```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
df_country.shape
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
#gives index of
```

```python
country_value=final_df.Country.value_counts().values
plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2F%%') #for top 3 countries only
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

CONCLUSIONS AND OBSERVSTIONS
1. From 4.5+ rating is excellent
2. From 4.0 and 4.4, rating is very good
3. From 3.5 and 3.9, rating is good
4. From 2.5 and 3.4, rating is average
5. From 1.8 and 2.4, rating is poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['brown','red','orange','yellow','green','green','dark green'])
```

```python
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['brown','red','orange','yellow','green','green','dark green'])
```
